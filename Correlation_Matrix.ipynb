```
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```


dat = pd.read_csv('top_3_each_team.csv')
dat = dat.loc[:,dat.columns != 'Unnamed: 1']
dat = dat.loc[:,dat.columns != 'Team']
col = list(dat.columns)
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
X_std = stdsc.fit_transform(dat[col].iloc[:,range(0,14)].values)
cov_mat = np.cov(X_std.T)
plt.figure(figsize=(10,10))
sn.set(font_scale=1.5)
hm = sn.heatmap(cov_mat,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 12},
                 cmap='coolwarm',                 
                 yticklabels=col,
                 xticklabels=col)
plt.title('Covariance matrix showing correlation coefficients', size = 18)
plt.tight_layout()
plt.show()

corrMatrix = dat.corr()
plt.figure(figsize = (10,10))
sn.set(font_scale=1.5)
new = sn.heatmap(corrMatrix, 
                 cbar=True,
                 fmt='.2f',
                 cmap='coolwarm',
                 yticklabels=col,
                 xticklabels=col)
plt.title('Correlation matrix', size = 18)
plt.tight_layout()
plt.show()
```
